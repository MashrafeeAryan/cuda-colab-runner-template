This part tells Colab how to compile your CUDA code.

You only need to change 3 things:

1. `executableName`
   This is the name of the program you want to create after compiling.

2. `includePath`
   This is the folder where your `.hpp` or `.h` header files are stored.
   Most C++ projects use `"include"`.

3. `sourceFiles`
   These are the `.cu` and `.cpp` files that must be compiled together.
   Add every file that contains real code your program needs.

For example, if your CUDA program uses:

* one CUDA file
* one C++ helper file
* one benchmark/main file

then all of those files should be listed inside `sourceFiles`.

After you set these values, the notebook automatically creates the `nvcc` compile command for you.


````python
# Name of the output executable
executableName = "benchmark_matrix_cpu_vs_cuda"

# Include folder for header files
includePath = "include"

# Source files needed to compile the program
sourceFiles = [
    "src/cuda/CUDA_MatrixMult.cu",
    "src/matrix/Matrix.cpp",
    "benchmarks/benchmark_matrix_cpu_vs_cuda.cu"
]

# Build the nvcc compile command automatically
compileCommand = f"nvcc -I {includePath} {' '.join(sourceFiles)} -o {executableName}"

print("Compile command:")
print(compileCommand)
````

````python
# Compile CUDA code
!{compileCommand}

# Run the executable
!./{executableName}
````
